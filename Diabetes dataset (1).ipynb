```python
import matplotlib.pyplot as plt

plt.hist(df["Age"], bins=20, color="skyblue", edgecolor="black")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['Glucose'], bins=30, kde=True, color='blue')  
plt.xlabel("Glucose Levels")
plt.ylabel("Frequency")
plt.title("Distribution of Glucose Levels")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(x=df["BMI"], color="orange")
plt.xlabel("BMI")
plt.title("Box Plot of BMI to Detect Outliers")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.scatterplot(x=df["Age"], y=df["Glucose"], color="blue")
plt.xlabel("Age")
plt.ylabel("Glucose")
plt.title("Scatter Plot of Age vs Glucose")
plt.show()
```
